### Data remarks
Normally, `pd.to_datetime()` find the good format, but  
  
If date in week mode, "2020-W07" :  
`Data['Date'] = pd.to_datetime(Data['Date'] + '-1',format = "%Y-W%W-%w")`  
  
But the format = "%Y-W%W-%w" don't respect ISO values and give error counting weeks!  
`Data['First Delivery'] = Data['Delivery week'].apply(lambda x:
datetime.datetime.strptime(x + '-0', "%G-W%V-%w"))`  
(slower, but can use %G and %V whom respect ISO)  
  
* the -1 is to choose monday as the day of the week by -%w

## Aggregate, filter, transform, groupby!

Think about the result you want, and work back from there. If you want to get a single value for each group, use aggregate() (or one of its shortcuts). If you want to get a subset of the original rows, use filter(). And if you want to get a new value for each original row, use transform().

### Here, we want only to manage the data by day, so, we will aggregate by summing repeated entries by days and by States! (and for time profiting, we will forget the NaN countries) 

```python
col = {"Lat": np.mean, "Long": np.mean, "recovereds": sum}
df_recovered = df_recovered.groupby(['Country', "Date"], as_index=False).agg(col)

col = {"Lat": np.mean, "Long": np.mean, "Deaths": sum}
df_death = df_death.groupby(['Country', "Date"], as_index=False).agg(col)

col = {"Lat": np.mean, "Long": np.mean, "confirmeds": sum}
df_confirmed = df_confirmed.groupby(['Country', "Date"], as_index=False).agg(col)
```
